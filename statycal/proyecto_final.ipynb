```python
#aca hice basicamente : Traeme las pinturas (matplotlib), los pinceles especiales (seaborn), y la calculadora (numpy) porque quiero dibujar un mapa visual."

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#latitud y longitud le digo donde estan las cordenadas del mapa 
#con esta funcion le digo que quiero que tenga un tamaño de 10 de ancho por 8 de alto."
plt.figure(figsize=(10,8))

sns.kdeplot(
    data=data,
    x="longitude",
    y="latitude",
    weights=np.log1p(data['price']),
    fill=True,
    cmap="viridis",
    levels=100,
    alpha=0.7
)
plt.title("Mapa de calor geográfico (escala de precios) - Airbnb Madrid")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.tight_layout()
plt.show()
```


- **`data=data`**: Usamos nuestra tabla de datos.
- **`x="longitude"`** y **`y="latitude"`**: Son como las coordenadas de un mapa. Le decimos dónde está cada alojamiento.
- **`weights=np.log1p(data['price'])`**: Pintamos zonas más caras con colores más intensos.  
  - **`log1p`** sirve para suavizar los valores (que no sean tan extremos).
- **`fill=True`**: Rellena el dibujo (no solo las líneas).
- **`cmap="viridis"`**: Usamos la paleta de colores violeta-verde.
- **`levels=100`**: Más detalle (muchos niveles de color en el mapa).
- **`alpha=0.7`**: Le damos transparencia media para que no quede muy cargado.

🔵 Colores:
Tonos oscuros/violetas = zonas con menor densidad de precios altos
Tonos verdes/amarillos = zonas con más concentración de alojamientos caros

Líneas (contornos):

Son curvas de nivel, como en los mapas topográficos.

Cada línea rodea áreas con valores similares: más juntas → más cambios bruscos.

📌 ¿Qué me dice este gráfico?
La zona más brillante (verde/amarilla) es el centro de Madrid, donde se concentran los alojamientos más caros.

Las zonas más oscuras tienen alojamientos más baratos o menos densidad.

Las formas del gráfico te muestran cómo se distribuyen espacialmente los precios, revelando barrios más exclusivos.

```python


# --- CARGAR ARCHIVOS ---
host = pd.read_parquet("airbnb_madrid_host.parquet")
property_data = pd.read_parquet("airbnb_madrid_property.parquet")
conditions = pd.read_parquet("airbnb_madrid_conditions.parquet")

# --- UNIR DATASETS ---
df = property_data.merge(host[["id", "host_is_superhost"]], on="id", how="inner")
df = df.merge(conditions[["id", "price"]], on="id", how="left")

# --- LIMPIEZA SUAVE ---
df = df[df["price"].notna()]               # Que tengan precio
df = df[df["price"] > 5]                    # Que el precio sea mayor a 5 €
df = df[df["host_is_superhost"].isin(["t", "f", True, False])]  # Que tengan estado válido

# --- MAPEAMOS SUPERHOST ---
df["host_is_superhost"] = df["host_is_superhost"].map({
    't': 'Sí', 'f': 'No', True: 'Sí', False: 'No'
})



# --- GRAFICO INTERACTIVO (Plotly) ---
fig = px.box(
    df,
    x="host_is_superhost",
    y="price",
    title='¿Ser Superhost permite cobrar más?',
    labels={"host_is_superhost": "¿Es Superhost?", "price": "Precio por noche (€)"},
    points="all",  # Mostrar también los puntos
    color="host_is_superhost",
    color_discrete_map={
        "Sí": "#FFB6C1",  # Rosado pastel
        "No": "#ADD8E6"   # Celeste pastel
    },
    template="plotly_white"
)

fig.update_layout(
    title_font_color="#6A5ACD",
    xaxis_title_font_color="#6A5ACD",
    yaxis_title_font_color="#6A5ACD"
)
fig.show()
```

```python


# --- Crear una nueva categoría de tiempo de respuesta ---
def categorizar_respuesta(x):
    if pd.isna(x):
        return "Sin dato"
    elif "within an hour" in x.lower():
        return "<1 hora"
    elif "within a few hours" in x.lower():
        return "<2-3 horas"
    elif "within a day" in x.lower():
        return "En el día"
    elif "a few days or more" in x.lower():
        return ">1 día"
    else:
        return "Otro"

# Aplicar la función
data["response_time_category"] = data["host_response_time"].apply(categorizar_respuesta)

# --- Análisis: Promedio de número de reviews según tiempo de respuesta ---
response_reviews = data.groupby("response_time_category")["number_of_reviews"].mean().sort_values(ascending=False)

import plotly.express as px

# --- Gráfico (interactivo pastel rosado y gris) ---
fig = px.bar(
    response_reviews.reset_index(),
    x="response_time_category",
    y="number_of_reviews",
    title="Promedio de reseñas según tiempo de respuesta del anfitrión",
    color="response_time_category",
    color_discrete_sequence=["#FFB6C1", "#FFC0CB", "#D8BFD8", "#C0C0C0", "#D3D3D3"],  # Rosados y grises
    template="plotly_white"
)

fig.update_layout(
    xaxis_title="Tiempo de respuesta",
    yaxis_title="Promedio de número de reseñas",
    title_font_color="#6A5ACD",
    xaxis_title_font_color="#6A5ACD",
    yaxis_title_font_color="#6A5ACD"
)

fig.show()
```
